````python
# Set Up 

import pandas as pd
import numpy as np
import seaborn as sns
````

````python
# Exercise 6.1
# creats a dataframe with the following data
# features contain the floats
# labels contain the integers 
# each value should be a list of four or five elements

df = pd.DataFrame({
    'features': [.2, -1.1, 2.6, 5.4],
    'labels': [1, 1, 0, 2]
})
df
````

````python
# Exercise 6.2
# Rename the labels column in df to label 

df.rename(columns={'labels': 'label'}, inplace=True)
df
````

````python
# Exercise 6.3
# Import the iris dataset from Seaborn and then sort by species in descending order

iris = sns.load_dataset('iris')
iris.sort_values(by='species', ascending=False)
iris
````

````python
# Exercise 6.4
# Use .value_counts() to find out how many records of each species the dataset has. 

iris['species'].value_counts().to_frame('count')
````

````python
# Exercise 6.5
# Print the number of records in 'iris' where the petal length <= 1.4 or petal length >= 1.6

len(iris[(iris['petal_length'] <= 1.4) | (iris['petal_length'] >= 1.6)])
````

````python
# Exercise 6.6
# Use .apply() to append a new column that displays the minimum of (petal_length, petal_width) for each observation
# Display the head and tail of the dataframe

iris['min_petal'] = iris.apply(lambda x: min(x['petal_length'], x['petal_width']), axis=1)
iris.head()
# iris.tail()
````

````python
iris.tail()
````

````python
# Exercise 6.7
# Use a pivot table to display the median and mean on sepal_length and sepal_width for each species

iris.pivot_table(index='species', values=['sepal_length', 'sepal_width'], aggfunc=[np.mean, np.median])
````
